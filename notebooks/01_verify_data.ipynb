```python
import pandas as pd
from sqlalchemy import create_engine, text

engine = create_engine("sqlite:///../data/fred_data.db")
```

```python
query = """
SELECT COUNT(*) AS total_rows,
       MIN(date) AS earliest_date,
       MAX(date) AS latest_date
FROM macro_series;
"""
pd.read_sql(text(query), con=engine.connect())
```

```python
query = """
SELECT *
FROM macro_series
ORDER BY date DESC
LIMIT 10;
"""
pd.read_sql(text(query), con=engine.connect())
```

```python
query = """
SELECT
    COUNT(*) AS total_rows,
    SUM(CASE WHEN fed_funds_rate IS NULL THEN 1 ELSE 0 END) AS missing_fed_funds,
    SUM(CASE WHEN consumer_credit IS NULL THEN 1 ELSE 0 END) AS missing_consumer_credit,
    SUM(CASE WHEN commercial_loans IS NULL THEN 1 ELSE 0 END) AS missing_commercial_loans,
    SUM(CASE WHEN mortgage_rate_30yr IS NULL THEN 1 ELSE 0 END) AS missing_mortgage_rate
FROM macro_series;
"""
pd.read_sql(text(query), con=engine.connect())
```

```python
query = """
SELECT
    CAST(SUBSTR(date, 1, 3) AS INTEGER) * 10 AS decade,
    ROUND(AVG(fed_funds_rate), 2) AS avg_rate,
    ROUND(MAX(fed_funds_rate), 2) AS peak_rate,
    ROUND(MIN(fed_funds_rate), 2) AS trough_rate
FROM macro_series
WHERE fed_funds_rate IS NOT NULL
GROUP BY decade
ORDER BY decade;
"""
pd.read_sql(text(query), con=engine.connect())
```
